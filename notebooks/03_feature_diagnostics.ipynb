# Feature Diagnostics

Inspect feature-store coverage, missingness, and key proposal figures without loading raw WRDS files.

```python
from pathlib import Path
import json

import pandas as pd

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent


def read_json(path):
    return json.loads((ROOT / path).read_text(encoding="utf-8"))


def maybe_read_csv(path, **kwargs):
    p = ROOT / path
    if not p.exists():
        return pd.DataFrame()
    return pd.read_csv(p, **kwargs)

ROOT
```

## Feature Store Manifest

```python
feature_store = read_json("artifacts/manifests/full_feature_store.json")
feature_store.get("diagnostics", feature_store)
```
