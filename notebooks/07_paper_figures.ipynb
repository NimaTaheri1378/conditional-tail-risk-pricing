# Paper Figures

Assemble the figure gallery for README, slides, and later manuscript use. Manuscript drafting is intentionally out of scope here.

```python
from pathlib import Path
import json

import pandas as pd

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent


def read_json(path):
    return json.loads((ROOT / path).read_text(encoding="utf-8"))


def maybe_read_csv(path, **kwargs):
    p = ROOT / path
    if not p.exists():
        return pd.DataFrame()
    return pd.read_csv(p, **kwargs)

ROOT
```

## Main Figures

```python
from IPython.display import Image, display
figures = [
    "cumulative_returns.png",
    "drawdown.png",
    "rolling_ic.png",
    "model_leaderboard.png",
    "benchmark_reconciliation.png",
    "alpha_bar.png",
    "robustness_forest.png",
    "signal_robustness_forest.png",
    "raw_chain_validation.png",
]
for name in figures:
    path = ROOT / "artifacts" / "figures" / "full" / name
    if path.exists():
        print(name)
        display(Image(filename=str(path)))
```
